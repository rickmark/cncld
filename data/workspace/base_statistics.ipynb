```python
%%sql
WITH base_stats AS (
SELECT
    COUNT(*) FILTER (WHERE canceled) AS canceled_count,
    COUNT(*) FILTER (WHERE revocable AND canceled) AS revocable_count,
    AVG(confidence) AS average_confidence,
    COUNT(*) AS total_count
FROM the_list),
    expanded_stats AS (
SELECT s.*,
       (CAST(s.canceled_count AS FLOAT) / s.total_count) * 100 AS cancel_ratio,
       (CAST(s.revocable_count AS FLOAT) / s.canceled_count) * 100 AS revocable_ratio
FROM base_stats AS s)
SELECT * FROM expanded_stats
```
